```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Step 1: Load dataset
df = pd.read_csv('../data/global_remote_work_wellbeing.csv')
print("Columns in dataset:")
print(df.columns)
print("\nSample data:")
print(df.head())

# Step 2: Simple Linear Regression with 'Screen_Time' to predict 'Stress_Level'
X = df[['Screen_Time']]
y = df['Stress_Level']

model = LinearRegression()
model.fit(X, y)

print("\nSimple Linear Regression Results:")
print(f"Intercept: {model.intercept_}")
print(f"Coefficient for Screen_Time: {model.coef_[0]}")

# Step 3: Plot regression line with scatter
plt.scatter(df['Screen_Time'], df['Stress_Level'], color='blue', alpha=0.5)
X_vals = np.array(df['Screen_Time'])
y_pred = model.intercept_ + model.coef_[0] * X_vals
plt.plot(X_vals, y_pred, color='red')
plt.xlabel('Screen Time')
plt.ylabel('Stress Level')
plt.title('Stress Level vs Screen Time with Regression Line')
plt.show()

# Step 4: Multiple Linear Regression with more features
features = ['Screen_Time', 'Daily_Working_Hours', 'Meetings_Attended', 'Emails_Sent', 
            'Physical_Activity_Steps', 'Sleep_Duration', 'Work_Life_Balance_Satisfaction']

X_multi = df[features]
y = df['Stress_Level']

model_multi = LinearRegression()
model_multi.fit(X_multi, y)

print("\nMultiple Linear Regression Results:")
print(f"Intercept: {model_multi.intercept_}")
print("Coefficients:")
for feat, coef in zip(features, model_multi.coef_):
    print(f"  {feat}: {coef}")

# Step 5: Evaluate the model performance
y_pred_multi = model_multi.predict(X_multi)
mse = mean_squared_error(y, y_pred_multi)
r2 = r2_score(y, y_pred_multi)

print(f"\nEvaluation Metrics for Multiple Regression:")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R^2 Score: {r2}")
```
